```python
wiki_df = pd.read_csv('wiki_df.csv')
postcode_geo = pd.read_csv('Geospatial_Coordinates.csv')
postcode_geo['Postal Code'] = postcode_geo['Postal Code'].str.strip()
postcode_geo = postcode_geo.rename({'Postal Code':'Postcode'},axis=1)
wiki_geo_df = pd.merge(wiki_df,postcode_geo,on='Postcode')
wiki_geo_df.head()
```

```python
# create map of New York using latitude and longitude values

map_toronto = folium.Map(location=[43.6532, -79.3832], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(wiki_geo_df['Latitude'], wiki_geo_df['Longitude'], wiki_geo_df['Borough'], wiki_geo_df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_toronto)
map_toronto
```
